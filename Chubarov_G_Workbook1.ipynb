```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
  return np.sqrt(1 + np.exp(np.sqrt(x)) + pow(np.cos(x), 2)) / abs(1 - pow(np.sin(x), 3)) + np.log(abs(2*x))

MIN = 10
MAX = 20

values = [function(x) for x in range(MIN, MAX)]
first_half = values[0:(MAX-MIN)//2]

print("Linear graph:")

figure, axis = plt.subplots(nrows=2, ncols=1)
figure.suptitle("Graphs:")

axis[0].plot(range(MIN, MAX), values)

print("Dots graph:")
axis[1].scatter(range(MIN, MIN + (MAX-MIN)//2), first_half)

```

```python
import numpy as np
import matplotlib.pyplot as plt

apple_data = [131.96, 121.26, 122.15, 131.46, 124.61, 136.96, 145.86, 151.83, 141.50, 149.80, 165.30, 177.57]
microsoft_data = [231.96, 232.38, 235.77, 252.18, 249.68, 270.90, 284.91, 301.88, 281.92, 331.62, 330.59, 336.32]
google_data = [91.787, 101.843, 103.432, 120.506, 120.578, 125.316, 135.221, 145.462, 133.266, 148.270, 142.452, 144.680]

months = np.arange(1, 13)

plt.figure(figsize=(8, 6), dpi=80)
plt.plot(months, apple_data, 'r', label="APPL")
plt.plot(months, microsoft_data, 'b', label="MSFT")
plt.plot(months, google_data, 'g', label="GOOG")
plt.xticks(months)
plt.legend(loc="upper left")
plt.show()
```

```python
import numpy as np

while True:
  print("Please, enter sign of an operation or 0 to stop execution")
  print("Additional functions:")
  print("1 - exp(x + y)")
  print("2 - sin(x + y)")
  print("3 - cos(x + y)")
  print("4 - x ^ y")
  sign = input()

  if sign == '0':
    break

  print("Please, enter x:")
  x = float(input())
  print("Please, enter y:")
  y = float(input())

  if sign == '+':
    result = x + y
  elif sign == '-':
    result = x - y
  elif sign == '*':
    result = x * y
  elif sign == "/":
    result = x / y
  elif sign == '1':
    result = np.exp(x + y)
  elif sign == '2':
    result = np.sin(x + y)
  elif sign == '3':
    result = np.cos(x + y)
  elif sign == '4':
    result = pow(x, y)
  print("Result: ", result)
  print("-----")
  
```
